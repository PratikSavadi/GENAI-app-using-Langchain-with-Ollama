OLLAMA is platfrom where we can use open source LLM models like LLAMA3,PHI3,Mstral and GAMMA.We can use all these LLM models
in our local machine without using API KEY where OPENAI is LLM model we use with API KEY

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
#ollama by default uses llma2 embeddings
embeddings=(OllamaEmbeddings(model="gemma2")
           )
```

```python
embeddings
```

```python
l1=embeddings.embed_documents(
    ['My Name is Pratik',
    'I m learning GEN AI',]
)
```
